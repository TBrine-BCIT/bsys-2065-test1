# The random module
- Produces "random" numbers
- Random number generators are based on a deterministic algorithm (pseudo-random generators)

```python
import random

prob = random.random()       # in the range [0.0, 1.0) – not including 1.0
print(prob)

result = prob * 5      # in the range [0.0, 5.0) – not including 5.0
print(result)

diceThrow = random.randrange(1, 7)       # return an int, one of 1,2,3,4,5,6
print(diceThrow)
```

## Conditional Execution: Binary Selection
- Selection statements (conditional statements) check conditions and change the behavior of the program accordingly
- if statement (binary selection) - two possible paths of execution
<br />
<br />
if BOOLEAN EXPRESSION:
<br />
&nbsp;&nbsp;&nbsp;&nbsp;STATEMENTS_1        # executed if BOOLEAN EXPRESSION condition evaluates to True
<br />
else:
<br />
&nbsp;&nbsp;&nbsp;&nbsp;STATEMENTS_2        # executed if BOOLEAN EXPRESSION condition evaluates to False

```python
x = 14
if x % 2 == 0:
    print(x, "is even")
else:
    print(x, "is odd")
```

## Omitting the else Clause: Unary Selection
- Unary selection - the else clause is omitted entirely

```python
x = -10
if x < 0:
    print("The negative number ",  x, " is not valid here.")
print("This is always printed")
```

## Nested conditionals

```python
x = 10
y = 10
if x < y:
    print("x is less than y")
else:
    if x > y:
        print("x is greater than y")
    else:
        print("x and y must be equal")
```

## Chained conditionals

```python
x = 10
y = 10
if x < y:
    print("x is less than y")
elif x > y:
    print("x is greater than y")
else:
    print("x and y must be equal")
```

- elif is an abbreviation of else if
- There is no limit of the number of elif statements
- Only a single (and optional) final else statement is allowed and it must be the last branch in the statement

## Boolean Functions
- Boolean functions return Boolean values

```python
def isDivisible(x, y):
    if x % y == 0:        # or use return x % y == 0
        result = True
    else:
        result = False
    return result

print(isDivisible(10, 5)) 

if isDivisible(10, 5):
    print("# do something ...")
else:
    print("# do something else ...")
```
