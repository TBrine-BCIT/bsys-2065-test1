## Turtle Graphics

## Instances
- An instance is an independent object
- Each instance has its own attributes and methods
- As long as they have different names, you can operate them independently

```python
import turtle

#Create two turtle instances
tanner = turtle.Turtle()
hannah = turtle.Turtle()

# Create a screen for our turtles to move around on
wn = turtle.Screen()

# Set the color and shape of our turtles
tanner.color("green", "gold")
tanner.shape("circle")
hannah.color("purple", "blue")
hannah.shape("turtle")

# Move our turtles around
tanner.backward(100)
hannah.left(90)
hannah.forward(200)

# This is for my code to work in Jupyter, you won't need this bit.
turtle.done()

try:
    turtle.bye()
except:
    print("bye")
```

## The for Loop
- A basic building block of many programs is to be able to repeat some code over and over again – iteration
- for statement implements iteration

```python
for invitee in ["Joe", "Amy", "Brad", "Angelina", "Zuki", "Thandi", "Paris"]:
    print("Hi", invitee+ ". Please come to my party on Saturday!")
    print("It's going to be so fun!")
print("I really hope everyone comes")
```

- name – loop variable
- [  ] – list
- print statement – loop body – indented – use 4 spaces
- terminating condition

```python
import turtle            # set up alex
wn = turtle.Screen()
alex = turtle.Turtle()

for aColor in ["yellow", "red", "purple", "blue"]:
    alex.color(aColor)
    alex.forward(50)
    alex.(90)

turtle.done()

try:
    turtle.bye()
except:
    print("bye")
```

## The range Function
- range objects deliver a sequence of values to the for loop.
- Always starts with 0 – e.g. range(4) is 0, 1, 2, 3
- Computer scientists like to count from 0 - https://www.howtogeek.com/149225/why-do-computers-count-from-zero/ 
- range(start, stop, step) – stops 1 before the stop parameter value (stop-1), step is the number to count by

```python
for num in range(3, 19, 4):
    print(num)
```

```python
## Turtle
import turtle
wn = turtle.Screen()
wn.bgcolor("lightgreen")
tess = turtle.Turtle()
tess.color("blue")
tess.shape("turtle")

print(list(range(5, 60, 2)))
tess.up()                     # this is new
for size in range(5, 60, 2):    # start with size = 5 and grow by 2
    tess.stamp()                # leave an impression on the canvas
    tess.forward(size)          # move tess along
    tess.right(24)              # and turn her

turtle.done()

try:
    turtle.bye()
except:
    print("bye")
```
